# HEALTH SITUATION

```python
health_situation = pd.read_csv('covid19-in-india/HospitalBedsIndia.csv')
```

```python
health_situation.head(10)
```

```python
sns.heatmap(health_situation.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
health_situation['NumPrimaryHealthCenters_HMIS'] = health_situation['NumPrimaryHealthCenters_HMIS'].apply(lambda x: int(x))
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x = 'State/UT', y = 'NumPrimaryHealthCenters_HMIS', data = health_situation)
```

```python
health_situation.dropna()
```

```python
health_situation['NumCommunityHealthCenters_HMIS'] = health_situation['NumCommunityHealthCenters_HMIS'].apply(lambda x: int(x))
```

```python
health_situation['NumDistrictHospitals_HMIS'] = health_situation['NumDistrictHospitals_HMIS'].apply(lambda x: int(x))
```

```python
health_situation['TotalPublicHealthFacilities_HMIS'] = health_situation['TotalPublicHealthFacilities_HMIS'].apply(lambda x: int(x))
```

```python
health_situation['NumPublicBeds_HMIS'] = health_situation['NumPublicBeds_HMIS'].apply(lambda x: int(x))
```

```python
health_situation['NumRuralHospitals_NHP18'] = health_situation['NumRuralHospitals_NHP18'].apply(lambda x: int(x))
```

```python
health_situation['NumRuralBeds_NHP18'] = health_situation['NumRuralBeds_NHP18'].apply(lambda x: int(x))
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x = 'State/UT', y = 'NumCommunityHealthCenters_HMIS', data = health_situation)
```

```python
health_situation.corr()['NumPrimaryHealthCenters_HMIS'].sort_values()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(health_situation.corr())
```
